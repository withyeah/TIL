# `abs()`

> 절대값은 숫자(int, float)가 들어오면 절대값을 반환하고, 복소수(complex)가 들어오면 그 크기를 반환합니다.
> 
> `my_abs(x)`를 만들어보세요.

```python
# 예제 입력 및 출력
print(abs(3+4j), abs(-0.0), abs(-5))
```

```python
# 아래에 코드를 작성해주세요.
def my_abs(numbers):
    if type(numbers) == type(complex()):
        return ((numbers.real**2) + (numbers.imag**2))**0.5
    elif type(numbers) == type(float()):
        return (numbers**2)**0.5
    elif numbers < 0:
        return -int(numbers)
    elif numbers >= 0:
        return numbers
```

```python
# 강사님 답
def my_abs(x):
    if type(x) == complex:
        return (x.imag**2 + x.real**2)**0.5
    else:
        if x == 0:
            return x ** 2
        if x < 0:
            return x * -1
        else:
            return x
```

```python
# 결과 확인
print(my_abs(3+4j), my_abs(-0.0), my_abs(0.0), my_abs(-5), my_abs(3), my_abs(-0))
```

# all(x)

> all은 인자로 받는 iterable(range, list)의 모든 요소가 참이면(또는 iterable이 비어있으면) True, 거짓이 하나라도 있으면 False를 리턴한다.
>
>이와 같은 my_all(x)을 작성해보세요

```python
# 예제 입력 및 출력
print(all([1, 2, 5, '6']))
print(all([[], 2, 5, '6']))
print(all([]))
print(all([0]))
```

```python
# 아래에 코드를 작성해주세요.
def my_all(items):
    for item in items:
        if not item:
            return False
    return True
# 비어있으면 아예 for 문을 안돌기 때문에 return True를 수행
```

```python
# 결과 확인
print(my_all([1, 2, 5, '6']))
print(my_all([[], 2, 5, '6']))
print(my_all([]))
print(my_all([0]))
```

# any(x)

> any는 인자로 받는 iterable(range, list)의 하나라도 참이면 True를 리턴하고, 인자가 모두 거짓일 경우에만 False를 리턴한다. (all(x)의 반대.)
>
> 이와 같은 my_any(x)를 작성해보세요.

```python
# 예제 입력 및 출력
print(any([1, 2, 5, '6']))
print(any([[], 2, 5, '6']))
print(any([0]))
```

```python
# 아래에 코드를 작성해주세요.
def my_any(items):
    for item in items:
        if item:
            return True
    return False
```

```python
# 결과 확인
print(my_any([1, 2, 5, '6']))
print(my_any([[], 2, 5, '6']))
print(my_any([0]))
```

> 이와 같은 `my_bin(x)`를 작성해보세요.

```python
# bin 예제 입력 및 출력
print(bin(4096), bin(5), bin(10))
```

```python
# 아래에 코드를 작성해주세요.
def my_bin(x):
    if x < 2:
        print(f'0b{x}', end='')
    else:
        my_bin(x//2)
        print(x%2, end='')

# def my_bin(x):
#     mb_str = ''
#     if x < 2:
#         mb_str = f'0b{x}' + mb_str
#     else:
#         my_bin(x//2)
#         mb_str = f'{x%2}' + mb_str 
#     return mb_str
        
my_bin(5)
# print(bin(4096))
# print(bin(5))
```

```python
def my_bin(x):    
    mb_str = ""
    while x//2 >= 1:
        remain = x%2
        x = x//2
        mb_str = str(remain) + mb_str
        if x < 2 :
            return '0b' + str(x) + mb_str
    return '0b' + mb_str + f'{x}' 

print(my_bin(4096), my_bin(5), my_bin(10), my_bin(1), my_bin(0))
```

```python
# my_bin 예제 입력 및 출력
print(my_bin(4096) == bin(4096))
print(my_bin(5) == bin(5))
```

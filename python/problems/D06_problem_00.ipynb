# 소대소대

> 단어의 짝수번째 알파벳은 대문자로, 홀수번째 알파벳은 소문자로 바꾼 문자열을 return 하는 함수 `up_and_low`을 작성하세요.

예시)
```python
up_and_low('applepie') #=> aPpLePiE
up_and_low('spaceship') #=> sPaCeShIp
```

````python
# 여기에 코드를 작성하세요.

# 내 답
def up_and_low(word):
    result = []
    for i in range(len(word)):
        if i%2 == 1:
            upper = word[i].upper()
            result.append(upper)
        else:
            lower = word[i].lower()
            result.append(lower)
    return ''.join(result)
````

````python
#강사님 답

def up_and_low(word):
    new_str = []
    for idx, w in enumerate(word):
        if idx % 2:
            new_str += w.upper()
        else:
            new_str += w.lower()
    return ''.join(new_str)
````

````python
# 강사님 답 -> list comprehension

def up_and_low(word):
    new_str = [w.upper() if idx%2 else w.lower() 
               for idx, w in enumerate(word)]
    return ''.join(new_str)
````

````python
# 해당 코드를 통해 올바른 결과가 나오는지 확인하세요.
print(up_and_low('applepie'))
print(up_and_low('spaceship'))
````

# 숫자 패턴  `복습하기!!!!!!!!!!!!!!!!!!!!!!!!!!!`

> 원하는 행까지 아래의 패턴을 생성하는 함수 `pattern`를 작성하세요. 만약 인자가 0이나 음의 정수인 경우 "" 즉, 빈 문자열로 반환하세요. 짝수가 인수로 전달되면 패턴은 통과된 짝수보다 작은 최대 홀수까지 계속되어야 합니다.

예시)

```python
pattern(9)

1
333
55555
7777777
999999999

pattern(6)

1
333
55555
```

주의)

- 패턴에 공백은 없습니다.
- 함수 내부에서 `print()`를 사용하는 것이 아니라, 하나의 문자열을 `return`하여서 출력하여야 합니다.

````python
# 여기에 코드를 작성하세요.
def pattern(n):
    string = ''
    a = n
    if a % 2 == 0:
        a -= 1
    for x in range(1, a+1):
        if x % 2 != 0:
            string += str(x) * x
            if x != a:
                string += '\n'
    return string
````

````python
def pattern(n):
    r = [str(x)*x for x in range(1, n+1) if x%2]  
    return '\n'.join(r)
````

````python
def pattern(n):
    r = [str(x)*x for x in range(1, n+1, 2)] # 간격 2로 줘서 홀수만 들어가게
    return '\n'.join(r)
````

````python
# 해당 코드를 통해 올바른 결과가 나오는지 확인하세요.
print(pattern(9))
print(pattern(6))
````

# 무엇이 중복일까

> 중복된 요소만 뽑아서 출력하는 `duplicated` 함수를 작성하시오.

예시)

```python
duplicated(['a', 'b', 'c', 'b', 'd', 'm', 'n', 'n', 'b']) #=> ['b', 'n']
```

````python
# 여기에 코드를 작성하세요.

# 내 답 (count와 set사용)

some_list = ['a', 'b', 'c', 'b', 'd', 'm', 'n', 'n']
def duplicated(some_list):
    duplicates = []
    for alphabet in some_list:
        if some_list.count(alphabet) > 1:
            duplicates.append(alphabet)
    return list(set(duplicates))
````

````python
# 강사님 답 (set 없이)

def duplicated(some_list):
    duplicates = []
    for value in some_list:
        if some_list.count(value) > 1:
            if value not in duplicates:    # set 대신 이미 duplicates에 있는지 if로 확인
                duplicates.append(value)
    return duplicates
````

````python
# 성원님 답 - list(set()) 두 번 사용해서 중복 제거

def duplicated(some_list):
    no_duplicate = list(set(some_list)
    for i in no_duplicate:
        some_list.remove(i)
    return list(set(some_list))
````
